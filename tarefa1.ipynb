```python
from scipy.stats import chi2_contingency



def bivariada_sequela(coluna):
    # display(df.groupby(coluna)['apresenta_alguma_sequela'].value_counts(normalize=True).unstack().T)
    display(df.groupby('apresenta_alguma_sequela')[coluna].value_counts().unstack().fillna(0))
    display(df.groupby(coluna)['apresenta_alguma_sequela'].value_counts(normalize=True).unstack().fillna(0).T)
    # display(chi2_contingency(df.groupby('apresenta_alguma_sequela')[coluna].value_counts().unstack()))

# bivariada_sequela('grau')
```

```python
categoricas = [c for c in  df.select_dtypes(include='object').columns.to_list() if 'dat' not in c and 'quando' not in c]

drop_this = ['IL-6 1:2 (16/06/2022)pg/ml', 'IL-6 1:2 (final)', 'TNF-α 1:2 (final)']

categoricas = [c for c in categoricas if c not in drop_this]
```

```python
columns_to_remove = ['ID']
for c in categoricas:
    
    print("*" * 50)
    print(c)
    bivariada_sequela(c)
```
